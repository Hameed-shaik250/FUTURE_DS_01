This project analyzes customer churn behavior using the Telco Customer Churn dataset.
The goal is to identify patterns and factors that influence customer churn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df['Churn'].value_counts()
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Contract Type vs Churn")
plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Internet Service vs Churn")
plt.show()
```

```python
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title("Payment Method vs Churn")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
plt.title("Senior Citizen vs Churn")
plt.show()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df[['tenure','MonthlyCharges','TotalCharges']].corr()
```

## Key Insights from Customer Churn Analysis

1. Month-to-month contract customers churn the most.
2. Fiber optic users have higher churn rates.
3. Customers with higher monthly charges churn more.
4. Long-tenure customers generate higher revenue.

## Conclusion

The analysis shows that contract type, monthly charges, and internet service type are major factors influencing customer churn.

Customers with month-to-month contracts and higher monthly charges tend to churn more frequently. In contrast, customers with longer tenure are less likely to leave the service.

Understanding these patterns can help companies design better customer retention strategies.
